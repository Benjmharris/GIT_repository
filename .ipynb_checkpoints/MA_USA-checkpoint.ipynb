```julia
using DataFrames, CSV, LinearAlgebra, BlackBoxOptim, Statistics
```

```julia
mutable struct DataTargets
    μ::Float64   # log income per capita
    V_y::Float64 # variance of log income per capita
    ρ::Float64   # IGM coefficient
end

d=DataTargets(NaN,NaN,NaN);
```

```julia
CountryData = CSV.read(".\\Data.csv");
```

```julia
size = length(CountryData.Code)
```

```julia
global betaoutput = ones(size, 7) * 666;
global parameters = ones(4) * 666;
```

```julia
global DataMatrix = [CountryData[:,2] CountryData[:,3] CountryData[:,4]];
```

```julia
global mean_μ = mean(CountryData[:,2]);
global std_μ = std(CountryData[:,2]);
global mean_V_y = mean(CountryData[:,3]);
global std_V_y = std(CountryData[:,3]);
global mean_ρ = mean(CountryData[:,4]);
global std_ρ = std(CountryData[:,4]);
global corr_μV_y = cor(CountryData[:,2], CountryData[:,3]);
global corr_V_yρ = cor(CountryData[:,3], CountryData[:,4]);
global corr_μρ = cor(CountryData[:,2], CountryData[:,4]);
```

```julia
corr_μV_y, corr_V_yρ, corr_μρ
```

```julia
global minV_ω = 0.05 * mean(CountryData[:,3]);
global maxV_ω = 5.0 * mean(CountryData[:,3]);
global minδ = (1+0.001)^30 - 1;
global maxδ = (1+0.05)^30 - 1;
global minα = 0.15;
global maxα = 0.85;
minV_ω, maxV_ω, minδ, maxδ, minα, maxα
```

```julia
function Va_function(beta_m,beta_a,params)
    βa=beta_a;βm=beta_m;
    α=params[1]*(maxα-minα)+minα;
    V_ω=params[2]*(maxV_ω-minV_ω)+minV_ω;
    δ=params[3]*(maxδ-minδ)+minδ;
    Z=1/params[4]-0.999;
    Va_val = V_y_function(βm,βa,params) .* (1 .- βa) ./ βa;
    return Va_val
end
```

```julia
function Vm_function(beta_m,beta_a,params)
    βa=beta_a;βm=beta_m;
    α=params[1]*(maxα-minα)+minα;
    V_ω=params[2]*(maxV_ω-minV_ω)+minV_ω;
    δ=params[3]*(maxδ-minδ)+minδ;
    Z=1/params[4]-0.999;
    Vm_val = ((α^2) .* βa .* Va_function(βm,βa,params) .+ V_ω) .* (1 .- βm) ./ βm;
    return Vm_val
end
```

```julia
function λ_function(beta_m,beta_a,params)
    βa=beta_a;βm=beta_m;
    α=params[1]*(maxα-minα)+minα;
    V_ω=params[2]*(maxV_ω-minV_ω)+minV_ω;
    δ=params[3]*(maxδ-minδ)+minδ;
    Z=1/params[4]-0.999;
    λ_val = α .* βm ./ (1 .+ δ .- α .* βa .* (1 .- βm));
    return λ_val
end
```

```julia
function μ_function(beta_m,beta_a,params)
    βa=beta_a;βm=beta_m;
    α=params[1]*(maxα-minα)+minα;
    V_ω=params[2]*(maxV_ω-minV_ω)+minV_ω;
    δ=params[3]*(maxδ-minδ)+minδ;
    Z=1/params[4]-0.999;
    μ_val = (1 ./  (1 .- α)) .* (log(Z) .+ α .* log.(λ_function(βm,βa,params)) .+ (1/2) .* (V_ω .- βm .* Vm_function(βm,βa,params)));
    return μ_val
end
```

```julia
function ρ_function(beta_m,beta_a,params)
    βa=beta_a;βm=beta_m;
    α=params[1]*(maxα-minα)+minα;
    V_ω=params[2]*(maxV_ω-minV_ω)+minV_ω;
    δ=params[3]*(maxδ-minδ)+minδ;
    Z=1/params[4]-0.999;
    ρ_val = α .* (βa .* (1 .- βm) .+ βm)
    return ρ_val
end
```

```julia
function V_y_function(beta_m,beta_a,params)
    βa=beta_a;βm=beta_m;
    α=params[1]*(maxα-minα)+minα;
    V_ω=params[2]*(maxV_ω-minV_ω)+minV_ω;
    δ=params[3]*(maxδ-minδ)+minδ;
    Z=1/params[4]-0.999;
    V_y_val = βm .* V_ω ./ (1 .- α .* ρ_function(βm,βa,params))
    return V_y_val
end
```

```julia
function correl_function(beta_m,beta_a,params)
    βa=beta_a;βm=beta_m;
    α=params[1]*(maxα-minα)+minα;
    V_ω=params[2]*(maxV_ω-minV_ω)+minV_ω;
    δ=params[3]*(maxδ-minδ)+minδ;
    Z=1/params[4]-0.999;
    μ=DataMatrix[:,1]; V_y=DataMatrix[:,2];ρ=DataMatrix[:,3];
    corr1 = cor(μ, μ_function(βm,βa,params))
    corr2 = cor(V_y, V_y_function(βm,βa,params))
    corr3 = cor(ρ, ρ_function(βm,βa,params))
    return [corr1 corr2 corr3]
end
```

```julia
function error_function(beta_m,beta_a,params)
    βa=beta_a;βm=beta_m;
    α=params[1]*(maxα-minα)+minα;
    V_ω=params[2]*(maxV_ω-minV_ω)+minV_ω;
    δ=params[3]*(maxδ-minδ)+minδ;
    Z=1/params[4]-0.999;
    correl = correl_function(βm,βa,params)
    m1 = 1 - correl[1]
    m2 = 1 - correl[2]
    m3 = 1 - correl[3]
    m4 = 1 - mean_μ/mean(μ_function(βm,βa,params))
    m5 = cor(μ_function(βm,βa,params), ρ_function(βm,βa,params)) - corr_μρ
    m6 = cor(V_y_function(βm,βa,params), ρ_function(βm,βa,params)) - corr_V_yρ
    return [m1 m2 m3 m4 m5 m6]
end
```

```julia
function norm_function(z)
    x = z[1:size]
    y = z[size+1:2*size]
    u = z[2*size+1:2*size+4]
    m = error_function(x,y,u)
    return sum(diag(m * W * m'))
end
```

```julia
global W=Diagonal([1.50875, 1.24125, 0.25, 1.0, 1.0, 1.0]); #Choose weights
global maxtimeval = 20000; #time limit for solver algorithm
W
```

```julia
result = bboptimize(norm_function; SearchRange = (0.0, 1.0), NumDimensions = 2*size+4, MaxTime = maxtimeval);
betas = best_candidate(result);
betaoutput[:,1] = betas[1:size];
betaoutput[:,2] = betas[size+1:2*size];
parameters = betas[2*size+1:2*size+4];
totalobjfuncvalue = best_fitness(result);
```

```julia
global α_val=parameters[1]*(maxα-minα)+minα;
global V_ω_val=parameters[2]*(maxV_ω-minV_ω)+minV_ω;
global δ_val=parameters[3]*(maxδ-minδ)+minδ;
global Z_val=1/parameters[4]-0.999;
global i_val = (1+δ_val)^(1/30)-1;
α_val, V_ω_val, δ_val, i_val, Z_val
```

```julia
cor(DataMatrix[:,1], μ_function(betaoutput[:,1], betaoutput[:,2], parameters)), cor(DataMatrix[:,2], V_y_function(betaoutput[:,1], betaoutput[:,2], parameters)), cor(DataMatrix[:,3], ρ_function(betaoutput[:,1], betaoutput[:,2], parameters)), cor(V_y_function(betaoutput[:,1], betaoutput[:,2], parameters), ρ_function(betaoutput[:,1], betaoutput[:,2], parameters)), cor(μ_function(betaoutput[:,1], betaoutput[:,2], parameters), ρ_function(betaoutput[:,1], betaoutput[:,2], parameters))
```

```julia
betaoutput[1,3] = α_val;
betaoutput[2,3] = V_ω_val;
betaoutput[3,3] = δ_val;
betaoutput[4,3] = Z_val;
betaoutput[5,3] = cor(DataMatrix[:,1], μ_function(betaoutput[:,1], betaoutput[:,2], parameters));
betaoutput[6,3] = cor(DataMatrix[:,2], V_y_function(betaoutput[:,1], betaoutput[:,2], parameters));
betaoutput[7,3] = cor(DataMatrix[:,3], ρ_function(betaoutput[:,1], betaoutput[:,2], parameters));
betaoutput[8,3] = mean(μ_function(betaoutput[:,1], betaoutput[:,2], parameters));
betaoutput[9,3] = cor(μ_function(betaoutput[:,1], betaoutput[:,2], parameters), ρ_function(betaoutput[:,1], betaoutput[:,2], parameters));
betaoutput[10,3] = cor(V_y_function(betaoutput[:,1], betaoutput[:,2], parameters), ρ_function(betaoutput[:,1], betaoutput[:,2], parameters));
betaoutput[11,3] = totalobjfuncvalue;
betaoutput[:,4] = μ_function(betaoutput[:,1], betaoutput[:,2], parameters);
betaoutput[:,5] = V_y_function(betaoutput[:,1], betaoutput[:,2], parameters);
betaoutput[:,6] = ρ_function(betaoutput[:,1], betaoutput[:,2], parameters);
betaoutput[:,7] = λ_function(betaoutput[:,1], betaoutput[:,2], parameters);
CSV.write("Beta.csv",  DataFrame(betaoutput), writeheader=false);
```
